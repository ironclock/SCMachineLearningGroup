```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("c:/Users/renee/desktop/project/SCL_adjusted_data.csv")
```

```python
data.head()
```

```python
adj=data.adjusted_data.values
```

```python
fig=plt.figure(figsize=(16,8))
iz=plt.plot(adj, color="red", linewidth=1, alpha=0.65)
plt.xlabel("Data Index", size=20)
plt.ylabel("mwh", size=20)
```

```python
idnan=np.array(data[data["adjusted_data"].isnull()].index)
print("There are ", idnan.size, "null values in the data.")
```

```python
idzero=np.where(adj==0)[0]
print("There are ", idzero.size, "zero values in the data.")
```

```python
idlarge=np.where(adj > 2000)[0]
print("There are ", idlarge.size, "large values in the data.")
```

```python
idlow=np.where(adj < 500)[0]
print("There are ", idlow.size, "low values in the data")
```
